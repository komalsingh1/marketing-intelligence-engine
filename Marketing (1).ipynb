```
%%sql -r dataframe_7
CREATE OR REPLACE VIEW STAGING.STG_CAMPAIGNS AS
SELECT
    CAMPAIGN_ID,
    UPPER(TRIM(CAMPAIGN_NAME))              AS CAMPAIGN_NAME,
    UPPER(TRIM(CHANNEL))                    AS CHANNEL,
    SPEND,
    START_DATE,
    END_DATE,
    DATEDIFF('day', START_DATE, END_DATE)   AS CAMPAIGN_DURATION_DAYS
FROM RAW.CAMPAIGNS
WHERE CAMPAIGN_ID IS NOT NULL
  AND SPEND > 0;
```

```
%%sql -r dataframe_8
CREATE OR REPLACE VIEW STAGING.STG_LEADS AS
SELECT
    LEAD_ID,
    CAMPAIGN_ID,
    UPPER(TRIM(LEAD_SOURCE))                AS LEAD_SOURCE,
    CREATED_AT,
    DATE(CREATED_AT)                        AS LEAD_DATE,
    UPPER(TRIM(COUNTRY))                    AS COUNTRY,
    UPPER(TRIM(INDUSTRY))                   AS INDUSTRY
FROM RAW.LEADS
WHERE LEAD_ID IS NOT NULL
  AND CAMPAIGN_ID IS NOT NULL;
```

```
%%sql -r dataframe_9
CREATE OR REPLACE VIEW STAGING.STG_CONVERSIONS AS
SELECT
    CONVERSION_ID,
    LEAD_ID,
    CONVERTED_AT,
    DATE(CONVERTED_AT)                              AS CONVERSION_DATE,
    DEAL_VALUE,
    UPPER(TRIM(PLAN_TYPE))                          AS PLAN_TYPE,
    CASE
        WHEN DEAL_VALUE >= 12000 THEN 'High'
        WHEN DEAL_VALUE BETWEEN 6000 AND 11999 THEN 'Mid'
        ELSE 'Low'
    END                                             AS DEAL_TIER
FROM RAW.CONVERSIONS
WHERE CONVERSION_ID IS NOT NULL
  AND DEAL_VALUE > 0;
```

```
CREATE OR REPLACE TABLE MART.CHANNEL_PERFORMANCE AS
SELECT
    c.CHANNEL,
    COUNT(DISTINCT l.LEAD_ID)                           AS TOTAL_LEADS,
    COUNT(DISTINCT cv.CONVERSION_ID)                    AS TOTAL_CONVERSIONS,
    ROUND(COUNT(DISTINCT cv.CONVERSION_ID) * 100.0 /
          NULLIF(COUNT(DISTINCT l.LEAD_ID), 0), 2)      AS CONVERSION_RATE_PCT,
    SUM(c.SPEND)                                        AS TOTAL_SPEND,
    COALESCE(SUM(cv.DEAL_VALUE), 0)                     AS TOTAL_REVENUE,
    COALESCE(ROUND(SUM(cv.DEAL_VALUE) /
          NULLIF(SUM(c.SPEND), 0), 2), 0)               AS REVENUE_ROI,
    ROUND(SUM(c.SPEND) /
          NULLIF(COUNT(DISTINCT cv.CONVERSION_ID), 0), 2) AS COST_PER_CONVERSION
FROM STAGING.STG_CAMPAIGNS c
LEFT JOIN STAGING.STG_LEADS l       ON c.CAMPAIGN_ID = l.CAMPAIGN_ID
LEFT JOIN STAGING.STG_CONVERSIONS cv ON l.LEAD_ID = cv.LEAD_ID
GROUP BY c.CHANNEL;
```

```
CREATE OR REPLACE TABLE MART.FUNNEL_ANALYSIS AS
WITH lead_counts AS (
    SELECT
        c.CHANNEL,
        COUNT(DISTINCT l.LEAD_ID)                       AS LEADS_GENERATED,
        COUNT(DISTINCT cv.CONVERSION_ID)                AS LEADS_CONVERTED,
        COUNT(DISTINCT l.LEAD_ID) -
            COUNT(DISTINCT cv.CONVERSION_ID)            AS LEADS_LOST
    FROM STAGING.STG_CAMPAIGNS c
    LEFT JOIN STAGING.STG_LEADS l        ON c.CAMPAIGN_ID = l.CAMPAIGN_ID
    LEFT JOIN STAGING.STG_CONVERSIONS cv ON l.LEAD_ID = cv.LEAD_ID
    GROUP BY c.CHANNEL
)
SELECT
    CHANNEL,
    LEADS_GENERATED,
    LEADS_CONVERTED,
    LEADS_LOST,
    ROUND(LEADS_CONVERTED * 100.0 /
          NULLIF(LEADS_GENERATED, 0), 2)                AS CONVERSION_RATE_PCT,
    ROUND(LEADS_LOST * 100.0 /
          NULLIF(LEADS_GENERATED, 0), 2)                AS DROP_OFF_RATE_PCT,
    -- Window function: rank channels by conversion rate
    RANK() OVER (ORDER BY LEADS_CONVERTED * 100.0 /
          NULLIF(LEADS_GENERATED, 0) DESC)              AS CONVERSION_RANK
FROM lead_counts;
```

```
CREATE OR REPLACE TABLE MART.CAMPAIGN_ROI AS
WITH campaign_metrics AS (
    SELECT
        c.CAMPAIGN_ID,
        c.CAMPAIGN_NAME,
        c.CHANNEL,
        c.SPEND,
        c.CAMPAIGN_DURATION_DAYS,
        COUNT(DISTINCT l.LEAD_ID)                           AS TOTAL_LEADS,
        COUNT(DISTINCT cv.CONVERSION_ID)                    AS TOTAL_CONVERSIONS,
        COALESCE(SUM(cv.DEAL_VALUE), 0)                     AS TOTAL_REVENUE
    FROM STAGING.STG_CAMPAIGNS c
    LEFT JOIN STAGING.STG_LEADS l        ON c.CAMPAIGN_ID = l.CAMPAIGN_ID
    LEFT JOIN STAGING.STG_CONVERSIONS cv ON l.LEAD_ID = cv.LEAD_ID
    GROUP BY c.CAMPAIGN_ID, c.CAMPAIGN_NAME, c.CHANNEL,
             c.SPEND, c.CAMPAIGN_DURATION_DAYS
)
SELECT
    CAMPAIGN_ID,
    CAMPAIGN_NAME,
    CHANNEL,
    SPEND,
    TOTAL_LEADS,
    TOTAL_CONVERSIONS,
    TOTAL_REVENUE,
    ROUND(TOTAL_REVENUE - SPEND, 2)                         AS NET_PROFIT,
    ROUND(TOTAL_REVENUE / NULLIF(SPEND, 0), 2)              AS ROI_MULTIPLIER,
    -- Window functions
    RANK() OVER (ORDER BY TOTAL_REVENUE DESC)               AS REVENUE_RANK,
    RANK() OVER (ORDER BY TOTAL_REVENUE /
                 NULLIF(SPEND, 0) DESC)                     AS ROI_RANK,
    ROUND(SUM(SPEND) OVER(), 2)                             AS TOTAL_BUDGET,
    ROUND(SPEND * 100.0 / SUM(SPEND) OVER(), 2)            AS BUDGET_SHARE_PCT
FROM campaign_metrics;
```

```
-- Best performing channels
SELECT * FROM MART.CHANNEL_PERFORMANCE
ORDER BY REVENUE_ROI DESC;

-- Funnel drop-off ranked
SELECT * FROM MART.FUNNEL_ANALYSIS
ORDER BY CONVERSION_RANK;

-- Top campaigns by ROI
SELECT * FROM MART.CAMPAIGN_ROI
ORDER BY ROI_RANK;
```

```
USE DATABASE MARKETING_ANALYTICS;
USE SCHEMA MART;

-- Monitoring view that flags anomalies
CREATE OR REPLACE VIEW MART.MARKETING_ALERTS AS

WITH channel_baselines AS (
    SELECT
        CHANNEL,
        TOTAL_LEADS,
        REVENUE_ROI,
        COST_PER_CONVERSION,
        -- Flag ROI below 2x
        CASE 
            WHEN REVENUE_ROI < 2 THEN TRUE 
            ELSE FALSE 
        END                                         AS LOW_ROI_FLAG,
        -- Flag leads spiking >50% above average across channels
        CASE 
            WHEN TOTAL_LEADS > AVG(TOTAL_LEADS) 
                 OVER() * 1.5 THEN TRUE 
            ELSE FALSE 
        END                                         AS LEAD_SPIKE_FLAG
    FROM MART.CHANNEL_PERFORMANCE
)
SELECT
    CHANNEL,
    TOTAL_LEADS,
    REVENUE_ROI,
    LOW_ROI_FLAG,
    LEAD_SPIKE_FLAG,
    CASE
        WHEN LOW_ROI_FLAG AND LEAD_SPIKE_FLAG 
            THEN '🚨 BOTH: Low ROI + Lead Spike'
        WHEN LOW_ROI_FLAG  
            THEN '⚠️ Low ROI Alert'
        WHEN LEAD_SPIKE_FLAG 
            THEN '📈 Lead Spike Alert'
        ELSE '✅ Healthy'
    END                                             AS ALERT_STATUS,
    CURRENT_TIMESTAMP()                             AS CHECKED_AT
FROM channel_baselines
WHERE LOW_ROI_FLAG = TRUE OR LEAD_SPIKE_FLAG = TRUE;
```

```
USE DATABASE MARKETING_ANALYTICS;
USE SCHEMA MART;

CREATE OR REPLACE TABLE MART.LEAD_QUALITY_SCORES AS

WITH lead_features AS (
    SELECT
        l.LEAD_ID,
        l.CAMPAIGN_ID,
        l.LEAD_SOURCE,
        l.COUNTRY,
        l.INDUSTRY,
        l.LEAD_DATE,
        cv.DEAL_VALUE,
        cv.PLAN_TYPE,
        cv.DEAL_TIER,
        CASE cv.LEAD_ID IS NOT NULL
            WHEN TRUE THEN 1 ELSE 0
        END                                         AS IS_CONVERTED,

        -- Score 1: Industry fit (Technology & Finance = high value for DocuSign-like SaaS)
        CASE l.INDUSTRY
            WHEN 'TECHNOLOGY' THEN 30
            WHEN 'FINANCE'    THEN 25
            WHEN 'HEALTHCARE' THEN 20
            WHEN 'RETAIL'     THEN 10
            ELSE 5
        END                                         AS INDUSTRY_SCORE,

        -- Score 2: Geography (enterprise markets)
        CASE l.COUNTRY
            WHEN 'USA'     THEN 30
            WHEN 'UK'      THEN 25
            WHEN 'GERMANY' THEN 20
            WHEN 'INDIA'   THEN 15
            ELSE 10
        END                                         AS COUNTRY_SCORE,

        -- Score 3: Plan type if converted
        CASE cv.PLAN_TYPE
            WHEN 'ENTERPRISE' THEN 30
            WHEN 'BUSINESS'   THEN 20
            WHEN 'STARTER'    THEN 10
            ELSE 0
        END                                         AS PLAN_SCORE,

        -- Score 4: Deal value tier
        CASE cv.DEAL_TIER
            WHEN 'High' THEN 10
            WHEN 'Mid'  THEN 5
            WHEN 'Low'  THEN 2
            ELSE 0
        END                                         AS DEAL_SCORE

    FROM STAGING.STG_LEADS l
    LEFT JOIN STAGING.STG_CONVERSIONS cv ON l.LEAD_ID = cv.LEAD_ID
),

scored AS (
    SELECT
        *,
        INDUSTRY_SCORE + COUNTRY_SCORE + 
        PLAN_SCORE + DEAL_SCORE               AS RAW_SCORE
    FROM lead_features
)

SELECT
    LEAD_ID,
    CAMPAIGN_ID,
    LEAD_SOURCE,
    COUNTRY,
    INDUSTRY,
    LEAD_DATE,
    IS_CONVERTED,
    PLAN_TYPE,
    DEAL_VALUE,
    INDUSTRY_SCORE,
    COUNTRY_SCORE,
    PLAN_SCORE,
    DEAL_SCORE,
    RAW_SCORE,

    -- Normalize to 0-100
    ROUND(RAW_SCORE * 100.0 / 100, 2)        AS QUALITY_SCORE,

    -- Bucket into tiers
    CASE
        WHEN RAW_SCORE >= 75 THEN 'A — Hot'
        WHEN RAW_SCORE >= 50 THEN 'B — Warm'
        WHEN RAW_SCORE >= 25 THEN 'C — Cold'
        ELSE 'D — Unqualified'
    END                                       AS LEAD_GRADE,

    -- Rank leads within each channel
    RANK() OVER (
        PARTITION BY LEAD_SOURCE 
        ORDER BY RAW_SCORE DESC
    )                                         AS RANK_IN_CHANNEL,

    -- Average score per channel using window function
    ROUND(AVG(RAW_SCORE) OVER (
        PARTITION BY LEAD_SOURCE
    ), 2)                                     AS AVG_CHANNEL_SCORE

FROM scored;
```

```
CREATE OR REPLACE TABLE MART.ATTRIBUTION_MODEL AS

WITH lead_journey AS (
    SELECT
        l.LEAD_ID,
        l.CAMPAIGN_ID,
        l.LEAD_SOURCE,
        l.CREATED_AT,
        cv.CONVERSION_ID,
        cv.DEAL_VALUE,
        cv.CONVERTED_AT,
        DATEDIFF('day', l.CREATED_AT, 
                 cv.CONVERTED_AT)            AS DAYS_TO_CONVERT,

        -- First touch = the channel that generated the lead
        FIRST_VALUE(l.LEAD_SOURCE) OVER (
            PARTITION BY l.LEAD_ID
            ORDER BY l.CREATED_AT
        )                                    AS FIRST_TOUCH_CHANNEL,

        -- Last touch = same in our dataset (single touch per lead)
        -- In multi-touch this would be the last interaction
        LAST_VALUE(l.LEAD_SOURCE) OVER (
            PARTITION BY l.LEAD_ID
            ORDER BY l.CREATED_AT
            ROWS BETWEEN UNBOUNDED PRECEDING 
                     AND UNBOUNDED FOLLOWING
        )                                    AS LAST_TOUCH_CHANNEL,

        -- Count touchpoints per lead (for linear split)
        COUNT(l.LEAD_ID) OVER (
            PARTITION BY l.LEAD_ID
        )                                    AS TOTAL_TOUCHPOINTS

    FROM STAGING.STG_LEADS l
    INNER JOIN STAGING.STG_CONVERSIONS cv ON l.LEAD_ID = cv.LEAD_ID
)

SELECT
    LEAD_ID,
    LEAD_SOURCE,
    FIRST_TOUCH_CHANNEL,
    LAST_TOUCH_CHANNEL,
    DEAL_VALUE,
    DAYS_TO_CONVERT,
    TOTAL_TOUCHPOINTS,

    -- First Touch Attribution: 100% credit to first channel
    ROUND(DEAL_VALUE, 2)                     AS FIRST_TOUCH_REVENUE,

    -- Last Touch Attribution: 100% credit to last channel
    ROUND(DEAL_VALUE, 2)                     AS LAST_TOUCH_REVENUE,

    -- Linear Attribution: split equally across touchpoints
    ROUND(DEAL_VALUE / 
          NULLIF(TOTAL_TOUCHPOINTS, 0), 2)   AS LINEAR_REVENUE,

    -- U-shaped: 40% first, 40% last, 20% middle
    ROUND(DEAL_VALUE * 0.4, 2)              AS USHAPED_FIRST_CREDIT,
    ROUND(DEAL_VALUE * 0.4, 2)              AS USHAPED_LAST_CREDIT,
    ROUND(DEAL_VALUE * 0.2 / 
          NULLIF(TOTAL_TOUCHPOINTS, 0), 2)   AS USHAPED_MIDDLE_CREDIT

FROM lead_journey;


-- Summary: compare models side by side per channel
CREATE OR REPLACE VIEW MART.ATTRIBUTION_COMPARISON AS
SELECT
    LEAD_SOURCE                              AS CHANNEL,
    COUNT(DISTINCT LEAD_ID)                  AS CONVERSIONS,
    SUM(FIRST_TOUCH_REVENUE)                 AS FIRST_TOUCH_TOTAL,
    SUM(LAST_TOUCH_REVENUE)                  AS LAST_TOUCH_TOTAL,
    SUM(LINEAR_REVENUE)                      AS LINEAR_TOTAL,
    SUM(USHAPED_FIRST_CREDIT + 
        USHAPED_LAST_CREDIT + 
        USHAPED_MIDDLE_CREDIT)               AS USHAPED_TOTAL
FROM MART.ATTRIBUTION_MODEL
GROUP BY LEAD_SOURCE
ORDER BY FIRST_TOUCH_TOTAL DESC;
```

```
CREATE OR REPLACE TABLE MART.COHORT_ANALYSIS AS

WITH cohorts AS (
    SELECT
        l.LEAD_ID,
        l.LEAD_SOURCE,
        l.INDUSTRY,
        -- Cohort = month lead was acquired
        DATE_TRUNC('month', l.CREATED_AT)    AS COHORT_MONTH,

        -- Conversion month
        DATE_TRUNC('month', 
            cv.CONVERTED_AT)                 AS CONVERSION_MONTH,

        cv.DEAL_VALUE,
        cv.PLAN_TYPE,

        -- Months between acquisition and conversion
        DATEDIFF('month',
            DATE_TRUNC('month', l.CREATED_AT),
            DATE_TRUNC('month', cv.CONVERTED_AT)
        )                                    AS MONTHS_TO_CONVERT,

        CASE WHEN cv.LEAD_ID IS NOT NULL 
             THEN 1 ELSE 0 
        END                                  AS IS_CONVERTED

    FROM STAGING.STG_LEADS l
    LEFT JOIN STAGING.STG_CONVERSIONS cv ON l.LEAD_ID = cv.LEAD_ID
)

SELECT
    COHORT_MONTH,
    LEAD_SOURCE,
    COUNT(DISTINCT LEAD_ID)                  AS COHORT_SIZE,
    SUM(IS_CONVERTED)                        AS CONVERTED,
    ROUND(SUM(IS_CONVERTED) * 100.0 /
          NULLIF(COUNT(DISTINCT LEAD_ID),0), 2) AS COHORT_CONVERSION_RATE,
    COALESCE(SUM(DEAL_VALUE), 0)             AS COHORT_REVENUE,
    ROUND(AVG(MONTHS_TO_CONVERT), 1)         AS AVG_MONTHS_TO_CONVERT,

    -- Running total of revenue across cohorts
    SUM(COALESCE(SUM(DEAL_VALUE), 0)) OVER (
        ORDER BY COHORT_MONTH
    )                                        AS CUMULATIVE_REVENUE,

    -- How this cohort compares to average conversion rate
    ROUND(SUM(IS_CONVERTED) * 100.0 /
          NULLIF(COUNT(DISTINCT LEAD_ID), 0)
          - AVG(SUM(IS_CONVERTED) * 100.0 /
          NULLIF(COUNT(DISTINCT LEAD_ID), 0)) 
          OVER(), 2)                         AS VS_AVG_CONVERSION_RATE

FROM cohorts
GROUP BY COHORT_MONTH, LEAD_SOURCE
ORDER BY COHORT_MONTH, LEAD_SOURCE;
```

```
USE DATABASE MARKETING_ANALYTICS;
USE SCHEMA MART;

CREATE OR REPLACE TABLE MART.CUSTOMER_LIFETIME_VALUE AS

WITH base AS (
    SELECT
        l.LEAD_ID,
        l.LEAD_SOURCE,
        l.INDUSTRY,
        l.COUNTRY,
        cv.PLAN_TYPE,
        cv.DEAL_VALUE,
        cv.CONVERTED_AT,
        c.SPEND,

        -- Average contract length assumption by plan (months)
        CASE cv.PLAN_TYPE
            WHEN 'ENTERPRISE' THEN 24
            WHEN 'BUSINESS'   THEN 18
            WHEN 'STARTER'    THEN 12
            ELSE 12
        END                                         AS CONTRACT_MONTHS,

        -- Assumed monthly churn rate by plan
        CASE cv.PLAN_TYPE
            WHEN 'ENTERPRISE' THEN 0.02   -- 2% churn
            WHEN 'BUSINESS'   THEN 0.04   -- 4% churn
            WHEN 'STARTER'    THEN 0.08   -- 8% churn
            ELSE 0.05
        END                                         AS MONTHLY_CHURN_RATE,

        -- Expansion revenue multiplier (upsell potential by industry)
        CASE l.INDUSTRY
            WHEN 'TECHNOLOGY' THEN 1.3
            WHEN 'FINANCE'    THEN 1.25
            WHEN 'HEALTHCARE' THEN 1.2
            WHEN 'RETAIL'     THEN 1.1
            ELSE 1.05
        END                                         AS EXPANSION_MULTIPLIER

    FROM STAGING.STG_LEADS l
    INNER JOIN STAGING.STG_CONVERSIONS cv ON l.LEAD_ID = cv.LEAD_ID
    INNER JOIN STAGING.STG_CAMPAIGNS c    ON l.CAMPAIGN_ID = c.CAMPAIGN_ID
),

clv_calc AS (
    SELECT
        *,
        -- Monthly recurring revenue
        ROUND(DEAL_VALUE / CONTRACT_MONTHS, 2)      AS MONTHLY_REVENUE,

        -- Average customer lifespan = 1 / churn rate (months)
        ROUND(1.0 / MONTHLY_CHURN_RATE, 1)          AS AVG_LIFESPAN_MONTHS,

        -- Base CLV = MRR * lifespan
        ROUND((DEAL_VALUE / CONTRACT_MONTHS) *
              (1.0 / MONTHLY_CHURN_RATE), 2)        AS BASE_CLV,

        -- Expanded CLV with upsell
        ROUND((DEAL_VALUE / CONTRACT_MONTHS) *
              (1.0 / MONTHLY_CHURN_RATE) *
              EXPANSION_MULTIPLIER, 2)              AS EXPANDED_CLV,

        -- CAC = spend attributed per conversion
        ROUND(SPEND / NULLIF(
            COUNT(*) OVER (PARTITION BY LEAD_SOURCE), 0
        ), 2)                                       AS ATTRIBUTED_CAC

    FROM base
)

SELECT
    LEAD_ID,
    LEAD_SOURCE,
    INDUSTRY,
    COUNTRY,
    PLAN_TYPE,
    DEAL_VALUE,
    MONTHLY_REVENUE,
    CONTRACT_MONTHS,
    MONTHLY_CHURN_RATE,
    AVG_LIFESPAN_MONTHS,
    BASE_CLV,
    EXPANDED_CLV,
    ATTRIBUTED_CAC,

    -- CLV to CAC ratio (>3 is healthy, >5 is excellent)
    ROUND(EXPANDED_CLV /
          NULLIF(ATTRIBUTED_CAC, 0), 2)             AS CLV_TO_CAC_RATIO,

    CASE
        WHEN EXPANDED_CLV / NULLIF(ATTRIBUTED_CAC, 0) >= 5
            THEN '🟢 Excellent'
        WHEN EXPANDED_CLV / NULLIF(ATTRIBUTED_CAC, 0) >= 3
            THEN '🟡 Healthy'
        WHEN EXPANDED_CLV / NULLIF(ATTRIBUTED_CAC, 0) >= 1
            THEN '🟠 Marginal'
        ELSE '🔴 Unprofitable'
    END                                             AS CLV_CAC_HEALTH,

    -- Rank by CLV within industry
    RANK() OVER (
        PARTITION BY INDUSTRY
        ORDER BY EXPANDED_CLV DESC
    )                                               AS CLV_RANK_IN_INDUSTRY,

    -- What % of total predicted revenue does this customer represent
    ROUND(EXPANDED_CLV * 100.0 /
          SUM(EXPANDED_CLV) OVER(), 2)              AS PCT_OF_TOTAL_CLV

FROM clv_calc;
```

```
CREATE OR REPLACE TABLE MART.CHANNEL_SATURATION AS

WITH channel_spend_buckets AS (
    SELECT
        c.CHANNEL,
        c.CAMPAIGN_ID,
        c.CAMPAIGN_NAME,
        c.SPEND,
        c.CAMPAIGN_DURATION_DAYS,
        COUNT(DISTINCT l.LEAD_ID)                   AS LEADS,
        COALESCE(SUM(cv.DEAL_VALUE), 0)             AS REVENUE,

        -- Spend per day (intensity)
        ROUND(c.SPEND /
              NULLIF(c.CAMPAIGN_DURATION_DAYS, 0), 2) AS DAILY_SPEND,

        -- Revenue per lead
        ROUND(COALESCE(SUM(cv.DEAL_VALUE), 0) /
              NULLIF(COUNT(DISTINCT l.LEAD_ID), 0), 2) AS REVENUE_PER_LEAD,

        -- Cost per lead
        ROUND(c.SPEND /
              NULLIF(COUNT(DISTINCT l.LEAD_ID), 0), 2) AS COST_PER_LEAD,

        -- ROI
        ROUND(COALESCE(SUM(cv.DEAL_VALUE), 0) /
              NULLIF(c.SPEND, 0), 2)                AS ROI

    FROM STAGING.STG_CAMPAIGNS c
    LEFT JOIN STAGING.STG_LEADS l        ON c.CAMPAIGN_ID = l.CAMPAIGN_ID
    LEFT JOIN STAGING.STG_CONVERSIONS cv ON l.LEAD_ID = cv.LEAD_ID
    GROUP BY c.CHANNEL, c.CAMPAIGN_ID,
             c.CAMPAIGN_NAME, c.SPEND,
             c.CAMPAIGN_DURATION_DAYS
),

channel_trends AS (
    SELECT
        *,
        -- Running total spend per channel
        SUM(SPEND) OVER (
            PARTITION BY CHANNEL
            ORDER BY SPEND
            ROWS BETWEEN UNBOUNDED PRECEDING AND CURRENT ROW
        )                                           AS CUMULATIVE_SPEND,

        -- ROI trend: compare to previous campaign in same channel
        LAG(ROI) OVER (
            PARTITION BY CHANNEL
            ORDER BY SPEND
        )                                           AS PREV_CAMPAIGN_ROI,

        -- Lead yield trend
        LAG(REVENUE_PER_LEAD) OVER (
            PARTITION BY CHANNEL
            ORDER BY SPEND
        )                                           AS PREV_REVENUE_PER_LEAD,

        -- Average ROI for the channel
        AVG(ROI) OVER (
            PARTITION BY CHANNEL
        )                                           AS AVG_CHANNEL_ROI,

        -- Spend rank within channel (1 = highest spend)
        RANK() OVER (
            PARTITION BY CHANNEL
            ORDER BY SPEND DESC
        )                                           AS SPEND_RANK

    FROM channel_spend_buckets
)

SELECT
    CHANNEL,
    CAMPAIGN_NAME,
    SPEND,
    DAILY_SPEND,
    LEADS,
    REVENUE,
    ROI,
    COST_PER_LEAD,
    REVENUE_PER_LEAD,
    CUMULATIVE_SPEND,
    AVG_CHANNEL_ROI,
    PREV_CAMPAIGN_ROI,

    -- ROI change vs previous campaign
    ROUND(ROI - COALESCE(PREV_CAMPAIGN_ROI, ROI), 2) AS ROI_DELTA,

    -- Diminishing return flag: ROI dropped AND spend increased
    CASE
        WHEN ROI < COALESCE(PREV_CAMPAIGN_ROI, ROI)
         AND SPEND > LAG(SPEND) OVER (
             PARTITION BY CHANNEL ORDER BY SPEND)
            THEN '⚠️ Diminishing Returns'
        WHEN ROI >= AVG_CHANNEL_ROI
            THEN '✅ Efficient'
        ELSE '📊 Monitor'
    END                                             AS SATURATION_STATUS,

    -- Efficiency score: ROI relative to channel average
    ROUND(ROI / NULLIF(AVG_CHANNEL_ROI, 0), 2)     AS EFFICIENCY_VS_AVG

FROM channel_trends
ORDER BY CHANNEL, SPEND;
```

```
CREATE OR REPLACE TABLE MART.LEAD_VELOCITY_RATE AS

WITH monthly_leads AS (
    SELECT
        DATE_TRUNC('month', LEAD_DATE)              AS MONTH,
        LEAD_SOURCE,
        COUNT(DISTINCT LEAD_ID)                     AS LEADS_THIS_MONTH,
        COUNT(DISTINCT CASE
            WHEN INDUSTRY IN ('TECHNOLOGY','FINANCE')
            THEN LEAD_ID END)                       AS QUALIFIED_LEADS
    FROM STAGING.STG_LEADS
    GROUP BY DATE_TRUNC('month', LEAD_DATE), LEAD_SOURCE
),

velocity AS (
    SELECT
        *,
        -- Previous month leads
        LAG(LEADS_THIS_MONTH) OVER (
            PARTITION BY LEAD_SOURCE
            ORDER BY MONTH
        )                                           AS PREV_MONTH_LEADS,

        LAG(QUALIFIED_LEADS) OVER (
            PARTITION BY LEAD_SOURCE
            ORDER BY MONTH
        )                                           AS PREV_QUALIFIED_LEADS,

        -- 3 month rolling average
        ROUND(AVG(LEADS_THIS_MONTH) OVER (
            PARTITION BY LEAD_SOURCE
            ORDER BY MONTH
            ROWS BETWEEN 2 PRECEDING AND CURRENT ROW
        ), 1)                                       AS ROLLING_3M_AVG,

        -- Cumulative leads per channel
        SUM(LEADS_THIS_MONTH) OVER (
            PARTITION BY LEAD_SOURCE
            ORDER BY MONTH
        )                                           AS CUMULATIVE_LEADS

    FROM monthly_leads
)

SELECT
    MONTH,
    LEAD_SOURCE,
    LEADS_THIS_MONTH,
    QUALIFIED_LEADS,
    PREV_MONTH_LEADS,
    ROLLING_3M_AVG,
    CUMULATIVE_LEADS,

    -- Lead Velocity Rate = MoM % change
    ROUND((LEADS_THIS_MONTH - COALESCE(PREV_MONTH_LEADS, LEADS_THIS_MONTH))
          * 100.0 /
          NULLIF(PREV_MONTH_LEADS, 0), 2)           AS LEAD_VELOCITY_RATE_PCT,

    -- Qualified lead velocity
    ROUND((QUALIFIED_LEADS - COALESCE(PREV_QUALIFIED_LEADS, QUALIFIED_LEADS))
          * 100.0 /
          NULLIF(PREV_QUALIFIED_LEADS, 0), 2)       AS QUALIFIED_LVR_PCT,

    -- Momentum signal
    CASE
        WHEN LEADS_THIS_MONTH > ROLLING_3M_AVG * 1.2
            THEN '🚀 Accelerating'
        WHEN LEADS_THIS_MONTH > ROLLING_3M_AVG
            THEN '📈 Growing'
        WHEN LEADS_THIS_MONTH = ROLLING_3M_AVG
            THEN '➡️ Stable'
        WHEN LEADS_THIS_MONTH < ROLLING_3M_AVG * 0.8
            THEN '📉 Declining'
        ELSE '⚠️ Slowing'
    END                                             AS MOMENTUM_SIGNAL,

    -- Is this channel above or below overall average velocity
    ROUND(LEADS_THIS_MONTH - AVG(LEADS_THIS_MONTH) OVER (
        PARTITION BY MONTH
    ), 1)                                           AS VS_MONTHLY_CHANNEL_AVG

FROM velocity
ORDER BY MONTH, LEAD_SOURCE;
```

```
CREATE OR REPLACE TABLE MART.TIME_DECAY_ATTRIBUTION AS

WITH lead_touches AS (
    SELECT
        l.LEAD_ID,
        l.LEAD_SOURCE,
        l.CAMPAIGN_ID,
        l.CREATED_AT                                AS TOUCH_TIME,
        cv.CONVERTED_AT,
        cv.DEAL_VALUE,

        -- Days between touch and conversion
        DATEDIFF('day', l.CREATED_AT,
                 cv.CONVERTED_AT)                   AS DAYS_BEFORE_CONVERSION,

        -- Total touches per lead
        COUNT(*) OVER (
            PARTITION BY l.LEAD_ID
        )                                           AS TOTAL_TOUCHES,

        -- Touch sequence number
        ROW_NUMBER() OVER (
            PARTITION BY l.LEAD_ID
            ORDER BY l.CREATED_AT
        )                                           AS TOUCH_NUMBER

    FROM STAGING.STG_LEADS l
    INNER JOIN STAGING.STG_CONVERSIONS cv ON l.LEAD_ID = cv.LEAD_ID
),

decay_weights AS (
    SELECT
        *,
        -- Time decay: half-life of 7 days
        -- Weight = 2^(-days/7) — closer touches get exponentially more credit
        ROUND(POWER(2, -1.0 * DAYS_BEFORE_CONVERSION / 7.0), 4) AS DECAY_WEIGHT,

        -- Sum of all decay weights for this lead (for normalization)
        SUM(POWER(2, -1.0 * DAYS_BEFORE_CONVERSION / 7.0)) OVER (
            PARTITION BY LEAD_ID
        )                                           AS TOTAL_DECAY_WEIGHT

    FROM lead_touches
),

normalized AS (
    SELECT
        *,
        -- Normalize weights so they sum to 1 per lead
        ROUND(DECAY_WEIGHT /
              NULLIF(TOTAL_DECAY_WEIGHT, 0), 4)     AS NORMALIZED_WEIGHT,

        -- Revenue attributed to this touch
        ROUND(DEAL_VALUE * (
            DECAY_WEIGHT / NULLIF(TOTAL_DECAY_WEIGHT, 0)
        ), 2)                                       AS ATTRIBUTED_REVENUE

    FROM decay_weights
)

SELECT
    LEAD_ID,
    LEAD_SOURCE,
    CAMPAIGN_ID,
    TOUCH_NUMBER,
    TOTAL_TOUCHES,
    TOUCH_TIME,
    CONVERTED_AT,
    DAYS_BEFORE_CONVERSION,
    DEAL_VALUE,
    DECAY_WEIGHT,
    NORMALIZED_WEIGHT,
    ATTRIBUTED_REVENUE,

    -- Channel summary using window
    SUM(ATTRIBUTED_REVENUE) OVER (
        PARTITION BY LEAD_SOURCE
    )                                               AS CHANNEL_TOTAL_ATTRIBUTED,

    ROUND(ATTRIBUTED_REVENUE * 100.0 /
          NULLIF(SUM(ATTRIBUTED_REVENUE) OVER(), 0), 2) AS PCT_OF_TOTAL_REVENUE

FROM normalized
ORDER BY LEAD_ID, TOUCH_NUMBER;


-- Summary view: compare time decay vs first/last touch per channel
CREATE OR REPLACE VIEW MART.ATTRIBUTION_FULL_COMPARISON AS
SELECT
    t.LEAD_SOURCE                               AS CHANNEL,
    COUNT(DISTINCT t.LEAD_ID)                   AS CONVERSIONS,
    SUM(t.ATTRIBUTED_REVENUE)                   AS TIME_DECAY_REVENUE,
    a.FIRST_TOUCH_TOTAL,
    a.LAST_TOUCH_TOTAL,
    a.LINEAR_TOTAL,
    -- Which model gives this channel the most credit
    CASE GREATEST(
        SUM(t.ATTRIBUTED_REVENUE),
        a.FIRST_TOUCH_TOTAL,
        a.LAST_TOUCH_TOTAL,
        a.LINEAR_TOTAL)
        WHEN SUM(t.ATTRIBUTED_REVENUE) THEN 'Time Decay'
        WHEN a.FIRST_TOUCH_TOTAL       THEN 'First Touch'
        WHEN a.LAST_TOUCH_TOTAL        THEN 'Last Touch'
        ELSE 'Linear'
    END                                         AS BEST_MODEL_FOR_CHANNEL
FROM MART.TIME_DECAY_ATTRIBUTION t
JOIN MART.ATTRIBUTION_COMPARISON a ON t.LEAD_SOURCE = a.CHANNEL
GROUP BY t.LEAD_SOURCE, a.FIRST_TOUCH_TOTAL,
         a.LAST_TOUCH_TOTAL, a.LINEAR_TOTAL
ORDER BY TIME_DECAY_REVENUE DESC;
```
